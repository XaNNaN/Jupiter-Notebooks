```python
import pandas as pd
import numpy as np

df = pd.read_csv("Task2.csv")
df = df[305: 320]
df = df.astype(np.float64)
df
```

```python
df.info()
```

```python
nat_cond = np.ones(15) - df.sum(axis=1).to_numpy()
nat_cond
```

```python
for i in range(0, 15):
    df.rename({i + 305 : i }, axis="index", inplace=True)

df
```

```python
for i in range(0, 15):
    df.iloc[i][i] = df.iloc[i][i] + nat_cond[i]

df
```

```python
finally_data = df.to_numpy()
```

```python
import matplotlib.pyplot as plt
import networkx as nx

DG = nx.DiGraph()  # создаём объект графа

# определяем список узлов (ID узлов)
nodes = [i for i in range(0, 15)]

# определяем список рёбер
# список кортежей, каждый из которых представляет ребро
# кортеж (id_1, id_2) означает, что узлы id_1 и id_2 соединены ребром
edges = [(i, y, df.iloc[i][y]) for  i in range(0, 15) for y in range(0, 15) if df.iloc[i][y] != 0]

# добавляем информацию в объект графа
DG.add_nodes_from(nodes)
DG.add_weighted_edges_from(edges)

# рисуем граф и отображаем его
nx.draw(DG, with_labels=True)
plt.show()
```

```python
nx.write_graphml_lxml(DG, "markov.graphml")
```

```python
def get_selv_vector(P, move_norm):
    delta = 0.00001
    
    P = P.to_numpy()
    P = P.T - np.eye(P.shape[0])
    
    B = np.zeros(P.shape[0] +1)
    B[-1] = 1 - move_norm
    N = np.ones(P.shape[0])
    D = [[0] for i in range(P.shape[0] +1)]

    P = np.vstack([P, N])
    P = np.hstack([P, D])
    
    solution = np.linalg.lstsq(P, B)
    print(sum(solution[0]))
    return solution[0]
    
```
